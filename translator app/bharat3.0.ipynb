```python
from tkinter import Tk, Label, Entry, Button, StringVar, OptionMenu, messagebox, Text
from googletrans import Translator

# Function to perform translation
def translate_text():
    input_text = input_entry.get()
    target_language = language_codes.get()

    translator = Translator()

    # Detect the language of the input text
    detected_lang = translator.detect(input_text)
    source_lang = detected_lang.lang

    try:
        # Translate the text
        translated_text = translator.translate(input_text, src=source_lang, dest=target_language)
        output_text.config(state="normal")
        output_text.delete("1.0", "end")
        output_text.insert("1.0", translated_text.text)
        output_text.config(state="disabled")
    except Exception as e:
        messagebox.showerror("Error", str(e))

# Create the main application window
root = Tk()
root.title("Language Translator")
root.geometry("500x500")

# Input text entry
input_label = Label(root, text="Enter the text to translate:")
input_label.pack()
input_entry = Entry(root, width=50)
input_entry.pack()

# Target language selection with language codes and names
language_codes = StringVar(root)
language_codes.set("fr")  # Default target language code ("fr" for French)
language_options = {
    "en": "English", "es": "Spanish", "fr": "French", "de": "German",
    "it": "Italian", "zh-CN": "Chinese (Simplified)", "zh-TW": "Chinese (Traditional)",
    "ja": "Japanese", "ko": "Korean", "ar": "Arabic", "ru": "Russian",
    "pt": "Portuguese", "hi": "Hindi", "bn": "Bengali", "ta": "Tamil", "ur": "Urdu"
}
language_menu = OptionMenu(root, language_codes, *language_options.keys())
language_menu.pack()

# Translate button
translate_button = Button(root, text="Translate", command=translate_text)
translate_button.pack()

# Output text with highlighting
output_text = Text(root, width=30, height=4, wrap="word", state="disabled")
output_text.pack()

# Note about language codes and names
note_label = Label(root, text="Note: Use language codes to select target language.\n"
                               "For example,\n"
                               "English: 'en'\n"
                               "Spanish: 'es'\n"
                               "French: 'fr'\n"
                               "German: 'de'\n"
                               "Italian: 'it'\n"
                               "Chinese (Simplified): 'zh-CN'\n"
                               "Chinese (Traditional): 'zh-TW'\n"
                               "Japanese: 'ja'\n"
                               "Korean: 'ko'\n"
                               "Arabic: 'ar'\n"
                               "Russian: 'ru'\n"
                               "Portuguese: 'pt'\n"
                               "Hindi: 'hi'\n"
                               "Bengali: 'bn'\n"
                               "Tamil: 'ta'\n"
                               "Urdu: 'ur'")
note_label.pack()

root.mainloop()
```
